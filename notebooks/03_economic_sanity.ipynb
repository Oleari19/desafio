```python
import pandas as pd

# Carregar a base
df_economic = pd.read_csv('../data/EconomicData_ZCTAs.csv')

# Exclusão da coluna Unnamed
df_economic.drop(columns=['Unnamed: 0'], inplace=True)

# Visualizar a descrição
display(df_economic.describe())

# Conferir tamanho (linhas x colunas)
print("\nTamanho:", df_economic.shape, "\n")

# Mostrar algumas tuplas para melhor entendimento
display(df_economic)

# Detalhar o tipo de cada atributo
print(df_economic.info())

# Definir a quantidade de valores nulos por atributo
print("\nNulos:\n")
print(df_economic.isnull().sum())
```

# Análise exploratória
- Dimensão geral da faixa de renda geral

```python
# Seleciona as colunas de renda (aquelas que começam com "TotalHouseholds_")
income_columns = [col for col in df_economic.columns if col.startswith('TotalHouseholds_')]

# Cria uma cópia só com os dados de renda
df_income = df_economic[income_columns].copy()

# Remove o prefixo "TotalHouseholds_" dos nomes das colunas para facilitar
df_income.columns = [col.replace('TotalHouseholds_', '') for col in income_columns]

# Soma total de domicílios por faixa de renda
total_per_income = df_income.sum()

# Calcula o percentual de cada faixa em relação ao total de domicílios
income_percentages = (total_per_income / total_per_income.sum()) * 100

# Exibe os percentuais ordenados (do maior para o menor)
income_percentages.sort_values(ascending=False).to_frame(name='Percentual (%)')
```

- Faixa de renda dominante das regiões

```python
# Seleciona colunas de renda
income_columns = [col for col in df_economic.columns if col.startswith('TotalHouseholds_')]

# Subset com os dados de renda e nome da região
df_income_region = df_economic[['Geographic Area Name'] + income_columns].copy()

# Identifica a faixa de renda dominante para cada região
dominant_income = df_income_region[income_columns].idxmax(axis=1)

# Remove prefixo "TotalHouseholds_"
dominant_income = dominant_income.str.replace('TotalHouseholds_', '', regex=False)

# Junta com o nome da região
df_result = pd.DataFrame({
    'ZCTA': df_economic['Geographic Area Name'],
    'Dominant income': dominant_income
})

# Exibe as primeiras linhas
display(df_result.head(10))
```

- ZCTAs mais ricas

```python
# Criar coluna com total de famílias com renda acima de 100 mil dólares
df_economic['HighIncomeHouseholds'] = (
    df_economic['TotalHouseholds_$100.000to$149.999'] +
    df_economic['TotalHouseholds_$150.000to$199.999'] +
    df_economic['TotalHouseholds_$200.000OrMore']
)

# Selecionar colunas para visualização
df_top_rich_zctas = df_economic[['Geographic Area Name', 'HighIncomeHouseholds']].copy()
df_top_rich_zctas = df_top_rich_zctas.rename(columns={
    'Geographic Area Name': 'ZCTA',
    'HighIncomeHouseholds': 'Famílias com renda > $100k'
})

# Ordenar e exibir as 10 ZCTAs com mais famílias ricas
df_top_rich_zctas = df_top_rich_zctas.sort_values(by='Famílias com renda > $100k', ascending=False)
display(df_top_rich_zctas.head(10))
```
