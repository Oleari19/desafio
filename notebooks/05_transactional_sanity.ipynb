# Tratamento e Informações Gerais
- Percebe-se que todas as informações ficaram em apenas uma coluna, portando ao carregar o arquivo já faz o tratamento para separar as colunas após o ";"

```python
# Carregar a base já com as colunas devidamente separadas
df_transactional = pd.read_csv('../data/transactional_data.csv', sep=';')

# Exclusão da coluna Unnamed
#df_transactional.drop(columns=['Unnamed: 0'], inplace=True)

# Visualizar a descrição
display(df_transactional.describe())

# Conferir tamanho (linhas x colunas)
print("\nTamanho:", df_transactional.shape, "\n")

# Mostrar algumas tuplas para melhor entendimento
display(df_transactional)

# Detalhar o tipo de cada atributo
print(df_transactional.info())

# Definir a quantidade de valores nulos por atributo
print("Nulos:\n")
print(df_transactional.isnull().sum())
```

# Qualidade
- Alterando o tipo da coluna "Testing Cost" de 'object' para 'float e substituindo as ',' por '.'
- Removendo os hórarios zerados da coluna "Date of birth"
- Alterando o tipo das colunas de data para o farmato date
- Adicionando uma coluna nova com a idade do paciente

```python
# Corrigindo a coluna 'Testing Cost'
df_transactional['Testing Cost'] = df_transactional['Testing Cost'].astype(str).str.replace(',', '.').str.strip()
df_transactional['Testing Cost'] = pd.to_numeric(df_transactional['Testing Cost'], errors='coerce')

# Converter ambas as colunas para datetime64
df_transactional['Date of birth'] = pd.to_datetime(df_transactional['Date of birth'], errors='coerce')
df_transactional['Date of service'] = pd.to_datetime(df_transactional['Date of service'], errors='coerce')

# Remover linhas com datas inválidas
df_transactional.dropna(subset=['Date of birth', 'Date of service'], inplace=True)

# Calcular idade no momento do serviço
df_transactional['Age at service'] = (
    (df_transactional['Date of service'] - df_transactional['Date of birth']).dt.days // 365
).astype(int)

# Visualizar resultados
df_transactional.info()
display(df_transactional[['Date of birth', 'Date of service', 'Age at service']].head())
```

# Análise Exploratória
- Analisar os exames mais realizados

```python
top_exams = df_transactional['CodItem'].value_counts().head(20)
top_exams.plot(kind='bar', figsize=(10,5), title='Top 20 exames mais realizados')
```

- Levando em consideração que: Testing Cost é o valor que o laboratório recebe pelo exame
- Analisar quais exames geram mais receita

```python
revenue_per_exam = df_transactional.groupby('CodItem')['Testing Cost'].sum().sort_values(ascending=False).head(20)
revenue_per_exam.plot(kind='bar', figsize=(10,5), title='Top 20 exames que mais geraram receita')
```

- Juntar os dois dados anteriomente analisados e ver a idade mais recorrente
- Top 10 exames por receita com a quantidade total de vezes realizados, e as 3 idade mais comuns deles com a quantidade de cada idade (ex: idade (x pessoas dessa idade).

```python
from collections import Counter

# Garantir que 'Testing Cost' é numérico
df_transactional['Testing Cost'] = df_transactional['Testing Cost'].astype(float)

# Calcular receita total por exame
revenue_by_exam = df_transactional.groupby('CodItem')['Testing Cost'].sum().sort_values(ascending=False)

# Pegar os 30 primeiros CodItems que mais geram receita
top30_coditems = revenue_by_exam.head(30).index

# Filtrar dados apenas desses 30 exames
top30_data = df_transactional[df_transactional['CodItem'].isin(top30_coditems)]

# Função para pegar as 3 idades mais comuns com quantidade
def top_3_ages(age_series):
    counter = Counter(age_series)
    top3 = counter.most_common(3)
    return ', '.join([f'{int(age)} (x{count})' for age, count in top3])

# Agrupar por CodItem com receita total, quantidade de exames e top 3 idades
top30_summary = top30_data.groupby('CodItem').agg(
    total_revenue=('Testing Cost', 'sum'),
    total_count=('Testing Cost', 'count'),
    top_3_common_ages=('Age at service', top_3_ages)
).sort_values(by='total_revenue', ascending=False)

display(top30_summary)
```
