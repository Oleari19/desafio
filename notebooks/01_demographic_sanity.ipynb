- Pré visualização

- Quantidade de valores nulos

```python
import pandas as pd

# Carregar a base
df_demographic = pd.read_csv('../data/DemographicData_ZCTAs.csv')

# Exclusão da coluna Unnamed
df_demographic.drop(columns=['Unnamed: 0'], inplace=True)

# Visualizar a descrição
display(df_demographic.describe())

# Conferir tamanho (linhas x colunas)
print("\nTamanho:", df_demographic.shape, "\n")

# Mostrar algumas tuplas para melhor entendimento
display(df_demographic)

# Detalhar o tipo de cada atributo
print(df_demographic.info())

# Definir a quantidade de valores nulos por atributo
print("\nNulos:\n")
print(df_demographic.isnull().sum())
```

# Tratamento de dados nulos
- Como o número de nulos é bem pequeno comparado ao número de registros, preencher com a mediana é o ideal, pois não vai gerar grandes impactos, uma vez que ela não é distorcida por valores absurdamente altos ou baixos.
- Nas coluna "MedianAgeInYears" e "SexRatio(males per 100 females)", os valores nulos foram substituídos pelo valor mediano das mesmas.
- O uso da mediana nesse contexto é importante para evitar distorção e para não perder informações
- Exclusão da coluna "Unnamed" 

```python
# Preencher os valores nulos com a mediana da coluna MedianAgeInYears
df_demographic['MedianAgeInYears'] = df_demographic['MedianAgeInYears'].fillna(df_demographic['MedianAgeInYears'].median())

# Preencher os valores nulos com a mediana da coluna SexRatio
df_demographic['SexRatio(males per 100 females)'] = df_demographic['SexRatio(males per 100 females)'].fillna(df_demographic['SexRatio(males per 100 females)'].median())

# Conferir se foram os dados nulos foram preenchidos
print("Nulos:\n")
print(df_demographic.isnull().sum())
```

# Análise exploratória
- Analisar as 10 ZCTAs mais populosas

```python
# Ordenar as 10 regiões com maior população
df_top_zctas = df_demographic.sort_values('TotalPopulation', ascending=False).head(10)
# Exibir a tabela com nome da região e população total
df_top_zctas[['GeographicAreaName', 'TotalPopulation']].reset_index(drop=True)
```

- Quais são as faixas-etárias mais dominantes das 10 ZCTAs mais populosas


```python
import matplotlib.pyplot as plt

# Ordenar as 10 regiões com maior população
df_top_zctas = df_demographic.sort_values('TotalPopulation', ascending=False).head(10)
# Exibir a tabela com nome da região e população total
df_top_zctas[['GeographicAreaName', 'TotalPopulation']].reset_index(drop=True)

# Colunas das faixas etárias
age_columns = [col for col in df_demographic.columns if 'Population_' in col]

# Dados com índice como nome da região
df_top_ages_zctas = df_top_zctas.set_index('GeographicAreaName')[age_columns]

# Convertendo para percentuais
df_ages_percent = df_top_ages_zctas.div(df_top_ages_zctas.sum(axis=1), axis=0) * 100

# Plot
plt.figure(figsize=(12, 6))
bottom = None

for col in age_columns:
    if bottom is None:
        plt.barh(df_ages_percent.index, df_ages_percent[col], label=col)
        bottom = df_ages_percent[col]
    else:
        plt.barh(df_ages_percent.index, df_ages_percent[col], left=bottom, label=col)
        bottom += df_ages_percent[col]

plt.title('Distribuição Percentual de Faixas Etárias nas 10 Regiões Mais Populosas')
plt.xlabel('Percentual da População (%)')
plt.ylabel('Região (ZCTA)')
plt.legend(title='Faixa Etária', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Colunas de faixas etárias
age_columns = [col for col in df_demographic.columns if 'Population_' in col]

# Seleciona as 10 ZCTAs mais populosas
df_top_zctas = df_demographic.sort_values('TotalPopulation', ascending=False).head(10)

# Dados de faixas etárias com índice sendo o nome da região
df_ages = df_top_zctas.set_index('GeographicAreaName')[age_columns]

# Converte para percentuais
df_percent = df_ages.div(df_ages.sum(axis=1), axis=0) * 100

# Pega as 3 faixas etárias dominantes para cada região
tabela_top3_faixas = []

for region in df_percent.index:
    top3 = df_percent.loc[region].sort_values(ascending=False).head(3)
    formatted = [f"{faixa.replace('Population_', '')} ({percent:.1f}%)" for faixa, percent in top3.items()]
    tabela_top3_faixas.append([region] + formatted)

# Criar DataFrame com os resultados
colunas = ['ZCTA', '1ª Faixa', '2ª Faixa', '3ª Faixa']
df_top3_faixas = pd.DataFrame(tabela_top3_faixas, columns=colunas)

# Exibir tabela
display(df_top3_faixas)
```
