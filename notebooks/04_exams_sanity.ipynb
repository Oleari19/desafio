```python
import pandas as pd

# Carregar a base
df_exams = pd.read_csv('../data/exams_data.csv')

# Visualizar a descrição
display(df_exams.describe())

# Conferir tamanho (linhas x colunas)
print("\nTamanho:", df_exams.shape, "\n")

# Mostrar algumas tuplas para melhor entendimento
display(df_exams)

# Detalhar o tipo de cada atributo
print(df_exams.info())

# Definir a quantidade de valores nulos por atributo
print("\nNulos:\n")
print(df_exams.isnull().sum())
```

# Análise exploratória
- Levando em consideração que: Testing Cost é o custo por exame, pois o mesmo CodItem na base transacional é mais caro, portanto o Testing Cost da base de exames é o custo do exame para o laboratório, já na base transacional, é o valor pago pelo paciente.
- Analisar os exames de maior custo

```python
df_exams.nlargest(10, 'Testing Cost')
```

- Analisar os exames de menor custo

```python
df_low_cost = df_exams.sort_values(by='Testing Cost', ascending=True).head(10)
df_low_cost
```
